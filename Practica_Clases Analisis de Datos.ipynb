```python
import pandas as pd

# 1. Volvemos a leer el archivo original para recuperar los datos
df = pd.read_excel('/tips.xlsx')

# 2. En lugar de borrar todo, borramos SOLO las columnas que están totalmente vacías
df = df.dropna(axis=1, how='all')

# 3. Si todavía ves NaNs, los llenamos con 0 para que no desaparezcan las filas
df = df.fillna(0)

print("¡Datos recuperados y limpios!")
display(df.head())
```
