```python
def actualizar_tabla_ventas(dataframe):
    # Lógica de cálculo rápida
    import numpy as np
    dataframe['Ventas'] = np.where(dataframe['Costo'] > 50000,
                                 dataframe['Costo'] * 1.30,
                                 dataframe['Costo'] * 1.10)

    # Recalcular Resultado
    dataframe['Resultado'] = dataframe['Ventas'] - dataframe['Costo']

    # Ordenar por importancia de vendedor
    dataframe['temp'] = dataframe.groupby('Vendedor')['Ventas'].transform('sum')
    dataframe.sort_values(by=['temp', 'Vendedor'], ascending=[False, True], inplace=True)

    # Limpieza final
    dataframe.drop(columns=['temp'], inplace=True)
    dataframe.reset_index(drop=True, inplace=True)
    return dataframe
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# La "Maquinaria" (Tu función Trigger)
#Aquí defines la lógica. Si mañana quieres cambiar los porcentajes, solo mueves los números aquí y vuelves a correr esta celda.
#python
def actualizar_todo(dataframe):
    # Cálculo de Ventas con lógica de Costo
    dataframe['Ventas'] = np.where(dataframe['Costo'] > 50000,
                                 dataframe['Costo'] * 1.30,
                                 dataframe['Costo'] * 1.10)

    # Cálculo de Óptimo y Margen
    dataframe['Optimo'] = dataframe['Ventas'] * 1.20
    dataframe['Marg_logr'] = ((dataframe['Ventas'] / dataframe['Optimo']) * 100).round(2)
    dataframe['Resultado'] = dataframe['Ventas'] - dataframe['Costo']

    # Ordenar por importancia de vendedor
    dataframe['temp'] = dataframe.groupby('Vendedor')['Ventas'].transform('sum')
    dataframe.sort_values(by=['temp', 'Vendedor'], ascending=[False, True], inplace=True)

    # Limpieza
    dataframe.drop(columns=['temp'], inplace=True)
    dataframe.reset_index(drop=True, inplace=True)
    return dataframe
```

```python
# A. Aplicamos todo el proceso a la tabla
df = actualizar_todo(df)

# B. Creamos la tabla resumen para la segunda hoja del Excel
df_resumen = df.groupby('Vendedor', as_index=False)['Ventas'].sum().sort_values(by='Ventas', ascending=False)

# C. Guardamos en Excel con dos hojas y descargamos
with pd.ExcelWriter('Informe_Final_Ventas.xlsx') as writer:
    df.to_excel(writer, sheet_name='Detalle_Filas', index=False)
    df_resumen.to_excel(writer, sheet_name='Resumen_Vendedores', index=False)

files.download('Informe_Final_Ventas.xlsx')
```
